```python
import pandas as pd
```

```python
url = 'https://www.fool.com/the-ascent/research/average-house-price-state/'
housing_cost_df = pd.read_html(url)[1]
```

```python
housing_cost_df.to_csv('housing_cost_per_state.csv', index=False)
housing_sorted = housing_cost_df.sort_values(by='Typical home value, June 2021', ascending=False)
housing_sorted = housing_sorted.reset_index(drop=True)
```

```python
income_by_state = pd.read_csv('\data\median income by state 2021.csv')  
income_by_state
```

```python
housing_merged = housing_cost_df.merge(income_by_state, how='left', on='State')
housing_merged
```

```python
housing_df = housing_merged.dropna()
housing_df
```

```python

#housing_df.astype({'Typical home value, June 2021': 'float'}).dtypes
housing_df['Typical home value, June 2021'] = housing_df['Typical home value, June 2021'].str.replace(',', '')
housing_df['Typical home value, June 2021'] = housing_df['Typical home value, June 2021'].str.replace('$', '')
housing_df['Typical home value, June 2021']
```

```python
housing_df['Typical home value, June 2021'] = pd.to_numeric(housing_df['Typical home value, June 2021'])
housing_df.dtypes
```

```python
percent_income = (housing_df['Typical home value, June 2021'] / housing_df['HouseholdIncome']) *100
percent_income
```

```python
housing_df.loc[:,'Percent House Cost Per Income (%)'] = percent_income
housing_df
```

```python
housing_df_2021 = housing_df.rename(columns={"Typical home value, June 2021": "Typical Home Value (June 2021)", "HouseholdIncome": "Household Income (2021)"})
housing_df_2021
```

```python
housing_df_2021.to_csv('housing cost per state.csv', index=False)
```
